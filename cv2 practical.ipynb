```
import cv2
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread('/content/th.jpg')
```

```
cv2_imshow(img)
```

```
height, width, _ = img.shape
```

```
for i in range(0,height-1):
  for j in range(0,width-1):
    pix=img[i,j]
    pix[0]=255-pix[0]
    pix[1]=255-pix[1]
    pix[2]=255-pix[2]
    img[i,j]=pix
cv2_imshow(img)
```
