```
#1
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt


sig= [np.sin(7*np.pi*t) for t in np.linspace(0,1, 1000)]

sig1= [np.sin(9*np.pi*t) for t in np.linspace(0,2, 1000)]

sig2= [np.sin(20*np.pi*t) for t in np.linspace(0,3, 1000)]

S=[i+J+K for i,J,K in zip(sig,sig1,sig2)]

plt.plot(S)

```

```
S1=np.fft.fft(S)
 
plt.plot(S1)
plt.show()
```
